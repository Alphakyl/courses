### The Story So Far 
***

**Normal Equations**: Solve $A^TA{\bf x} = A^T{\bf b}$

- $A^TA$ is $n \times n$
- $A^TA$ is nonsingular if cols of $A$ are linearly independent. 

**QR Factorization**: Find $Q$ and $R$ such that $A = QR$

- $Q$ is $m \times n$ with orthonormal columns
- $R$ is $n \times n$ upper triangular
- Solution is $\hat{\bf x} = R^{-1}Q^T{\bf b}$

### The Math Behind the  Bias-Variance Trade-Off 
***

Assume that there is a true relationship between response $y$ and data ${\bf X} = (X_1, X_2, \ldots, X_D)$ of the form 

$$
Y = f({\bf X}) + \epsilon 
$$

Here $f$ is a fixed (true) but unknown function and $\epsilon$ is a random error term independent of ${\bf X}$ with mean zero  

Function $f$ represents systematic information that ${\bf X}$ provides about $Y$ 

The error $\epsilon$ can come from different sources
- Features that influence $Y$ not included in our feature set 
- Measurements in data are not perfect


<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>



### The Math Behind the  Bias-Variance Trade-Off 
***

Our goal is to approximate the unknown function $f$ by a known function $\hat{f}$

Then, given some new test data, we can approximate 

$$
\hat{Y} = \hat{f}({\bf X})
$$

Think of $\hat{f}$ as a random variable itself.  The actual form depends on training it on some data 

Error in prediction can be measured by expected value of squared difference between $Y$ and $\hat{Y}$

$$
E\left[ (Y - \hat{Y})^2\right]
=
E\left[ (Y - \hat{f}({\bf X}))^2\right]
$$


<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>



### The Math Behind the  Bias-Variance Trade-Off 
***

We have 

$$
E\left[ \left( Y - \hat{f}(X) \right)^2 \right] = E\left[ \left( f\left(X\right) + \epsilon - \hat{f}\left( X \right) \right)^2 \right] 
= E\left[ \left( f\left(X\right) - \hat{f}\left( X \right) \right)^2 + 2\epsilon\left( f\left(X\right) - \hat{f}\left( X \right) \right) + \epsilon^2 \right] 
=  ~~~ E\left[~ (f\left(X\right) - \hat{f}\left( X \right))^2 \right] + Var\left(\epsilon \right) 
$$

Last step follows from $E[\epsilon] = 0$ and independence from ${\bf X}$

Decomposes error into *reducible* and *irreducible* 

<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>



### The Math Behind the  Bias-Variance Trade-Off 
***

Going further 

$$
~E\left[ \left( Y - \hat{f}(X) \right)^2 \right]
= 
E\left[~ (f\left(X\right) - \hat{f}\left( X \right))^2 \right] + Var\left(\epsilon \right)
$$

$$
= E[\left\{ (f(X)-E[\hat{f}(X)]) + (E[\hat{f}(X)] - \hat{f}(X) )\right\}^2] + V[\epsilon] 
$$


$$
= (f(X) - E[\hat{f}(X)])^2 + E[(~E[\hat{f}(X)] - \hat{f}(X)~)^2] + V[\epsilon] 
$$

$$
= \textrm{Bias}^2 + \textrm{Variance}^2 + V[\epsilon] 
$$

<br><br><br><br><br>
<br><br><br><br><br>
<br><br><br><br><br>

```python
sqbias = []
var = []
irred = []
mxDegree=10
numRuns=200
for degree in range(1,mxDegree,1):
    poly = PolynomialFeatures(degree=degree)
    meanfit = np.zeros(numTest)
    np.random.seed(0)
    for ii in range(numRuns):
        xTrain = np.linspace(-L,L,numTrain) + np.random.rand(numTrain) / (1.0*numTrain)
        yTrain = map(lambda x: truef(x), xTrain) + sig*np.random.randn(len(xTrain))
        ATrain = poly.fit_transform(xTrain.reshape(len(xTrain),1))
        ATest  = poly.fit_transform(xTest.reshape(len(xTest),1))
        beta, ssq = np.linalg.lstsq(ATrain, yTrain)[0:2]
        meanfit = meanfit + np.dot(ATest,beta)
    meanfit = meanfit / numRuns
    np.random.seed(0)
    varfit = np.zeros(numTest)
    for ii in range(numRuns):
        xTrain = np.linspace(-L,L,numTrain) + np.random.rand(numTrain) / (1.0*numTrain)
        yTrain = map(lambda x: truef(x), xTrain) + sig*np.random.randn(len(xTrain))
        ATrain = poly.fit_transform(xTrain.reshape(len(xTrain),1))
        ATest  = poly.fit_transform(xTest.reshape(len(xTest),1))
        beta, ssq = np.linalg.lstsq(ATrain, yTrain)[0:2]
        varfit = varfit + (np.dot(ATest,beta)-meanfit)**2
    varfit = varfit / numRuns
    sqbias.append(np.mean((meanfit-truef(xTest))**2))
    var.append(np.mean(varfit))
    irred.append(np.sum((yTest-truef(xTest))**2)/len(xTest))
    
testerror = np.array(irred)+np.array(var)+np.array(sqbias)
    
f, (ax1) = plt.subplots(1,1)
f.set_figheight(5)
f.set_figwidth(8)
ax1.set_xlabel("Model Complexity")
ax1.set_ylabel("Prediction Error")
ax1.plot(range(1,mxDegree,1),testerror,color="black",label="test error")
ax1.plot(range(1,mxDegree,1),np.array(sqbias),label="squared-bias")
ax1.plot(range(1,mxDegree,1),np.array(var),label="variance")
legend = ax1.legend(loc='upper right', shadow=True)
ax1.set_xlim([0,mxDegree]);
ax1.set_ylim([0,1]);
```
