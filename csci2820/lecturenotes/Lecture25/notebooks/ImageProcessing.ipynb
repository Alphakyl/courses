```python
import numpy as np 
#from matplotlib.pyplot import imshow, figure, axis, subplots, plot
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from PIL import Image, ImageDraw
```

```python
def inlineShow(img, sz=8):
    %matplotlib inline
    figure(figsize=(sz,sz))
    imshow(np.asarray(img), cmap = cm.Greys_r, interpolation='none')
    axis('off')
```

```python
im = np.loadtxt('gand.mat')
```

```python
im.shape
```

```python
inlineShow(im)
```

```python
U, s, V = np.linalg.svd(im, full_matrices=False)
```

```python
figure(figsize=(12,8))
plt.xlim(-5, 200)
plt.ylim(0, 1.1)
plt.plot(s[0:200]/s.max(),'ro')
plt.ylabel('singular vals');
```

```python
s5 = np.copy(s)
s5[5:] = 0.0
im5 = np.dot(U, np.dot(np.diag(s5), V))
inlineShow(im5)
```

```python
s10 = np.copy(s)
s10[10:] = 0.0
im10 = np.dot(U, np.dot(np.diag(s10), V))
inlineShow(im10)
```

```python
s25 = np.copy(s)
s25[25:] = 0.0
im25 = np.dot(U, np.dot(np.diag(s25), V))
inlineShow(im25)
```

```python
sumall = np.linalg.norm(s)
expvar = np.zeros(len(s))
for ii in np.arange(len(s)):
    expvar[ii] = np.linalg.norm(s[:ii])/sumall
    
cut999 = 0.999*np.ones(len(s))
cut9999 = 0.9999*np.ones(len(s))

figure(figsize=(12,8))
plt.xlim(0, len(s))
plt.ylim(.99, 1.0)
plt.plot(expvar,'r')
plt.plot(cut999,'k--')
plt.plot(cut9999,'k--')
plt.ylabel('explained variance');
```

```python
# Explained Variance = 0.99
np.sum(np.ones(len(expvar))[expvar<.99])
```
