#Lecture 10

***

<br>

Last time we talked about two very important subspaces associated with an $m \times n$ matrix $A$

* **The Column Space of A**: All linear combinations of columns of $A$. Subspace of $\mathbb{R}^m$. 
* **Important Fact**: The system $A\bx = {\bf b}$ is solvable only if ${\bf b}$ is in $C\myp{A}$


* **The Nullspace of A**: Space of all vectors $\bx$ such that $A\bx = {\bf 0}$.  Subspace of $\mathbb{R}^n$
* **Important Facts**: 
    1. $N(A)$ is characterized by the *special solutions* of $A\bx = {\bf 0}$. 
    2. The number of special solutions is the number of free variables 
    3. If $\bx$ and $\by$ are in $N(A)$ then any linear combo of $\bx$ and $\by$ is in $N(A)$




**Example 1**: Consider the matrix $A = \mymat{1 & 1 & 2 & 4 \\ 1 & 2 & 2 & 5 \\ 1 & 3 & 2 & 6}$

**Q**: Which of the columns of $A$ are important to the column space of $A$? 

**A**: Remember that $C(A)$ is all of the places in $\mathbb{R}^3$ we can get by taking linear combinations of the columns of $A$.  But sometimes, not every column of $A$ is actually necessary. Let's look at the columns of this particular matrix. 

**column 1**: $~~~\mymat{1 \\ 1 \\ 1}$ is important (mostly because it's the first column)

**column 2**: $~~~\mymat{1 \\ 2 \\ 3}$ is important because it's not parallel to column 1

**column 3**: $~~~\mymat{2 \\ 2 \\ 2}$ is 2 times column 1, so it doesn't provide any new information. 

**column 4**: $~~~\mymat{4 \\ 5 \\ 6}$ is 3 times column 1 plus column 2, so it also doesn't provide any new information.

So it turns out we can get anywhere in the column space of $A$ by taking linear combinations of columns 1 and 2.  In other words, columns 1 and 2 are important, and columns 3 and 4 are in some sense redundant.  

**Q**: OK, so what if it wasn't so easy to see?

**A**: Do elimination on $A$!

Let's put $A$ in upper echelon form: 

<br>

$$
A = 
\mymat{
1 & 1 & 2 & 4 \\ 
1 & 2 & 2 & 5 \\ 
1 & 3 & 2 & 6
}
~~ \sim ~~
\mymat{
1 & 1 & 2 & 4 \\ 
0 & 1 & 0 & 1 \\ 
0 & 2 & 0 & 2
}
~~ \sim ~~
\mymat{
1 & 1 & 2 & 4 \\ 
0 & 1 & 0 & 1 \\ 
0 & 0 & 0 & 0
}
= U 
$$

<br>

OK, so what does this tell us?  Notice that the reduced matrix has pivots in columns 1 and 2.  This tells us that columns 1 and 2 are important in the column space of $A$ (as we deduced above).  The columns corresponding to free variables (columns 3 and 4) are the redundant columns. 

If we want to know exactly how the redundant columns depend on the pivot columns it's helpful to further reduce the matrix to reduced row echelon form: 

<br> 

$$
U = 
\mymat{
1 & 1 & 2 & 4 \\ 
0 & 1 & 0 & 1 \\ 
0 & 0 & 0 & 0
}
~~ \sim ~~
\mymat{
1 & 0 & 2 & 3 \\ 
0 & 1 & 0 & 1 \\ 
0 & 0 & 0 & 0
}
= R 
$$

<br> 

From the reduced row echelon form $R$ we can immediately see the special solutions of $A\bx = {\bf 0}$: 

<br>

$$
\begin{array}{cccccc}
x_1 &  & +2x_3 & +3x_4 &=& 0  \\ 
   & x_2 &  & +x_4 &=& 0  \\
\end{array}
\quad \Rightarrow \quad 
\begin{array}{rcll}
x_1 &=& -2x_3  &- 3x_4 \\
x_2 &=& & -4x_4 \\ 
x_3 &=& x_3 &  \\
x_4 &=& & x_4 
\end{array}
\quad \Rightarrow \quad
\bx = x_3\mymat{-2 \\ 0 \\ 1 \\ 0} + x_4\mymat{-3 \\ -1 \\ 0 \\ 1}
$$

<br>

Recall that (col 3 = 2 col 1) and (col 4 = 3 col 1 + col 2), which is precisely what the special solutions tell us. 

**Summary**: To find which columns of $A$ are immportant for the columns space of $A$, and which columns are redundant you reduce $A$ to reduced row echelon form.  Then 

* The columns of $R$ with pivots tell you which columns of $A$ are important 
* The special solutions tell you how the redundant columns of $A$ depend on the important ones. 

<br> 

**New Key Term**: The **rank** of a matrix $A$ is the number of pivots in $U$ or $R$.  It's also the number of important columns of $A$ in the column space. We denote the rank of $A$ by $\textrm{rank}(A)$ or simply $r$. 

**Example 2**:  The rank of the matrix $A$ in the previous example is $r = 2$. 

Suppose that $A$ is $m \times n$ with rank $r$.  Then 

* $A$ has $r$ important columns 
* $A$ has $n-r$ free variables and $n-r$ special solutions. 

**Looking Ahead**: Later we'll refer to the rank of $A$ as the *dimension* of the column space of $A$.  The number of columns minus the rank of $A$ will be the *dimension* of the nullspace of $A$. 

With our new understanding of reduced forms, ranks, column spaces, and nullspaces, we're now ready to talk about all possible solutions of $A\bx = {\bf b}$ for a general (possibly non-square) matrix $A$. 

**Q**: Is it possible to find conditions on an arbitrary rhs vector ${\bf b}$ such that $A\bx = {\bf b}$ is solvable? 
**A**: Yes! 

**Example 3**: Consider solving $A\bx = {\bf b}$ for $A = \mymat{ 1 & 1 \\ 1 & 2 \\ -2 & -3}$

The ${\bf b}$'s for which the system are solvable are the ${\bf b}$'s in the column space of $A$.  But can we describe them explicitly for the given matrix $A$? Let's try to solve the system with a general rhs vector ${\bf b}$: 

<br>

$$
\myb{A ~ \left| ~ {\bf b}\right.} \quad = \quad 
\myaug{
1 & 1 && b_1 \\
1 & 2 && b_2 \\
-2 & -3 && b_3 
}
\quad \sim \quad 
\myaug{
1 & 1 && b_1 \\
0 & 1 && b_2 - b_1 \\
0 & -1 && b_3 + 2b_1 
}
\quad \sim \quad 
\myaug{
1 & 1 && b_1 \\
0 & 1 && b_2 - b_1 \\
0 & 0 && b_1 + b_2 + b_3 
}
$$

<br>

$$
\quad \sim \quad 
\myaug{
1 & 0 && 2b_2-b_1 \\
0 & 1 && b_2 - b_1 \\
0 & 0 && b_1 + b_2 + b_3 
}
$$

<br>

From the last row, we know that the system is solvable exactly when the rhs element is zero.  This means that ${\bf b}$ is in $C(A)$ exactly when $b_1 + b_2 + b_3 = 0$. 

**Example 4**: Write down the general or complete solution to $A\bx = {\bf b}$ for $A = \mymat{1 & 1 & 2 \\ 3 & 2 & 2}$ and ${\bf b} = \mymat{3 \\ 6}$. 

Row reducing we have 

<br>

$$
\myaug{
1 & 1 & 2 & 3 \\
3 & 2 & 2 & 6
}
\quad \sim \quad
\myaug{
1 & 1 & 2 & 3 \\
0 & -1 & -4 & -3
}
\quad \sim \quad
\myaug{
1 & 1 & 2 & 3 \\
0 & 1 & 4 & 3
}
\quad \sim \quad
\myaug{
1 & 0 & -2 & 0 \\
0 & 1 & 4 & 3
}
$$

<br>

Notice that the system has one free variable, and so it will have one special solution.  Writing the basic variables in terms of the free variables we have 

<br>

$$
\begin{array}{rcrr}
x_1 &=& 0 & +2x_3 \\
x_2 &=& 3 & -4x_3 \\
x_3 &=& 0 & +x_3 \\
\end{array}
\quad \Rightarrow \quad
\bx = 
\mymat{0 \\ 3 \\ 0} 
+ 
x_3 
\mymat{2 \\ -4 \\ 1}
$$

<br>

Let's check that for a given value of the free variable the solution actually works.  Let's make it easy and set $x_3 = 0$.  We then have

<br>

$$
\bx = \mymat{0 \\ 3 \\ 0}
$$

<br>

Then 

<br> 

$$
A\bx = 
\mymat{
1 & 1 & 2 \\
3 & 2 & 2
}
\mymat{ 0 \\ 3 \\ 0} 
= 
\mymat{0 + 3 + 0 \\ 0 + 6 + 0} 
= 
\mymat{3 \\ 6} 
\quad \checkmark
$$

<br> 

OK, so without the free variable term the constant vector in the solution, multiplied by $A$, produces the correct rhs.  But what about the vector associated with the free variable? 

<br> 

$$
A\bx = 
\mymat{
1 & 1 & 2 \\
3 & 2 & 2
}
\mymat{ 2 \\ -4 \\ 1} 
= 
\mymat{2 -4 + 2 \\ 6 - 8 + 2} 
= 
\mymat{0 \\ 0} 
$$

OK, so the constant vector in the complete solution gives us the rhs vector, and the vector associated with the free variable gets mapped to ${\bf 0}$ (and doesn't mess up the rhs).  

The constant vector that gives us the correct rhs vector is called the **particular solution** of $A\bx = {\bf b}$.  

The vector associated with the free variable is a special solution of $A\bx = {\bf 0}$ and is thus in the Nullspace of $A$.  We write the complete solution as follows 

<br> 

$$
\bx = \bx_p + \bx_n = \mymat{0 \\ 3 \\ 0} + x_3 \mymat{2 \\ -4 \\ 1}
$$

<br> 

**Example 5**: Find the complete solution to $A\bx = {\bf b}$ where $A = \mymat{1 & -2 & 3 \\ 1 & -2 & 3 \\ -2 & 4 & -6}$ and ${\bf b} = \mymat{2 \\ 2 \\ 4}$. 

Row reducing we have 

<br>

$$
\myaug{
1 & -2 & 3 & 2 \\
1 & -2 & 3 & 2 \\
-2 & 4 & -6 & 4 
}
\quad \sim \quad 
\myaug{
1 & -2 & 3 & 2 \\
0 &  0 & 0 & 0 \\
0 &  0 & 0 & 0 
}
$$

<br>

This time both $x_2$ and $x_3$ are free variables.  Writing $x_1$ in terms of $x_2$ and $x_3$ gives 

<br>

$$
x_1 = 2 + 2x_2 - 3x_3 
\quad \Rightarrow \quad
\begin{array}{rcrrr}
x_1 &=& 2 & +2x_2 & -3x_3 \\
x_2 &=& 0 & +x_2 & +0x_3 \\
x_3 &=& 0 & +0x_2 & +x_3
\end{array}
\quad \Rightarrow \quad
\bx = 
\mymat{2 \\ 0 \\ 0}
+ x_2 \mymat{2 \\ 1 \\ 0} 
+ x_3 \mymat{ -3 \\ 0 \\ 1}
$$

<br>

Then we have for the complete solution 

<br>

$$
\bx = \bx_p + \bx_n = \mymat{2 \\ 0 \\ 0} + \myp{ \quad x_2\mymat{2 \\ 1 \\ 0} + x_3\mymat{-3 \\ 0 \\ 1}\quad}
$$
