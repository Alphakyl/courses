**Example 1**: Consider the matrix $A = \mymat{1 & 2 \\ 2 & 4 \\ 4 & 8}$ in $\R{3\times 2}$. 

If in general $A$ is $m \times n$, then in this particular case we have $m=3$ and $n=2$.

We've already seen that we can find bases for the column space of $A$, $C(A)$, and the nullspace of $A$, $N(A)$, by performing row reduction on the matrix. 

<br>

$$
A \hs \sim \hs \mymat{1 & 2 \\ 0 & 0 \\ 0 & 0} \hs = \hs R 
$$

<br>

The first column of $R$ is a pivot column, and indicates that a basis for the column space of $A$ is the first column of $A$.  The second column of $R$ indicates that the variable $x_2$ is a free variable.  The special solution of $A\bx = {\bf 0}$ associated with $x_2$ will form a basis for the nullspace of $A$: 

<br>

$$
\begin{array}{rcr}
x_1 &=& -2x_2 \\
x_2 &=& x_2
\end{array}
\quad \Rightarrow \quad 
\bx = x_2\mymat{-2 \\ 1} 
$$

<br>

We then have the following representations of the column space and nullspace of $A$: 

<br>

$$
C(A) = \textrm{span}\mys{\hs\mymat{1 \\ 2 \\ 4}\hs}
\quad \textrm{and} \quad 
N(A) = \textrm{span}\mys{\hs\mymat{-2 \\ 1}\hs}
$$

<br>

Note that $C(A)$ is a subspace of $\R{m} = \R{3}$ and $N(A)$ is a subspace of $\R{n} = \R{2}$.  Furthermore, the rank of the matrix is $r = 1$. 

It turns out there are two other subspaces associated with the matrix $A$, the *row space* and *left nullspace* of $A$.   

The **row space** of $A$ is made up of all linear combinations of the *rows* of $A$.  Note that this is also the space of all linear combinations of the columns of $A^T$.  Then the row space of a matrix is the same as the column space of the matrix transpose.  Sometimes we will denote the row space of $A$ as $C(A^T)$. 

The left nullspace of $A$ is all vectors $\by$ such that $A^T\by = {\bf 0}$. So the left nullspace of $A$ is exactly the nullspace of $A^T$. The left nullspace gets it's name because if we take the transpose of both sides of the equation in the definition, we have 

<br>

$$
\myp{A^T\by}^T = {\bf 0}^T \quad \Rightarrow \quad \by^T A = {\bf 0}^T
$$

<br>

So the left nullspace is made up of all vectors $\by$ such that if we multiply $A$ on the left by the row-vector $\by^T$ we get the row vector ${\bf 0}^T$. 

Note that the row space, $C(A^T)$, is a subspace of $\R{n} = \R{2}$ and $N(A^T)$ is a subspace of $\R{m} = \R{3}$. 


These new spaces, along with the columnspace and nullspace of $A$, make up the so-called **four fundamental subspaces** associated with a matrix.  In particular, there are two fundamental subspaces associated with multiplication of $A$ by a vector, and two fundamental subspaces associated with multiplication of $A^T$ by a vector. 

OK, so how do we find bases for these new subspaces?  

**Row Space Basis**: It turns out that we get a basis for the row space for free if we've already done the row reduction to find bases for $C(A)$ and $N(A)$. 

Recall that after performing row reduction on $A$ to obtain $R$, the pivot columns indicate which columns of the original matrix $A$ that we should use for the basis for $C(A)$.  The reason we had to go back to $A$ to get the basis was because row reduction takes linear combinations of the rows of $A$.  After we've performed an elementary row operation, the columns of the reduced matrix are no longer in the column space of $A$ (remember, $C(A)$ is all linear combos of the columns of $A$). This is not a problem for the row space of $A$.  Since row reductions specifically takes linear combinations of the rows of $A$, the resulting rows of the reduced matrix *are* in the row space of $A$. This means that we can directly select the *pivot rows* of the reduced matrix $R$ as a basis for the $C(A^T)$. 

Recall we had 

<br>

$$
A \hs \sim \hs \mymat{1 & 2 \\ 0 & 0 \\ 0 & 0} \hs = \hs R 
$$

<br>

Since the only row of $R$ with a pivot is the first row, we have 

<br>

$$
C(A^T) = \textrm{span}\mys{\hs\mymat{1 \\ 2}\hs}
$$

<br>

Of course, there are many other (less easy) ways to find a basis for $C(A^T)$.  We could take the pivot rows of the original matrix $A$ (in this case, they're the same) or we could take the pivot rows of the upper echelon form matrix $U$. 

**Left Nullspace Basis**: Unfortunately, we cannot determine a basis for the left nullspace of $A$ from $R$.  To determine the left nullspace we have to perform row reduction on $A^T$.  For this particular example we have 

<br>

$$
A^T = \mymat{1 & 2 & 4 \\ 2 & 4 & 8} 
\hs \sim \hs 
\mymat{1 & 2 & 4 \\ 0 & 0 & 0}
$$

<br>

Here, the variables $y_2$ and $y_3$ are free, and will produce two special solutions to $A^T\by = {\bf 0}$.  We have 

<br>

$$
\begin{array}{rcrr}
y_1 &=& -2y_2 & -4y_3 \\
y_2 &=& y_2 & \\
y_3 &=&  & y_3
\end{array}
\quad \Rightarrow \quad 
\by = y_2 \mymat{-2 \\ 1 \\ 0} + y_3 \mymat{-4 \\ 0 \\ 1}
$$

<br>

Thus a the left nullspace of $A$ can be characterized as 

<br>

$$
N(A^T) = \textrm{span}\mys{\hs\mymat{-2 \\ 1 \\ 0}, \mymat{-4 \\ 0 \\ 1}\hs}
$$

<br>

To **summarize**: Bases for $C(A)$, $N(A)$, and $C(A^T)$ can be found by performing row reduction on $A$.  A basis for $N(A^T)$ can be found by performing row reduction on $A^T$. 

Let's see if we can say something about the dimensions of these spaces for an arbitrary matrix of size $m \times n$. 

First, notice that we found bases for $C(A)$ and $C(A^T)$ by identifying columns of $A$ and rows of $R$ that contained pivots after row reduction. Since the number of pivots after row reduction is the rank of a matrix, this means that the dimension of both $C(A)$ and $C(A^T)$ is always equal to $r$.  (Incidentally, we also just proved that the rank of a matrix and it's transpose are always the same.)

OK, what about the two nullspaces?  If $A$ is $m \times n$ then it has $n$ columns.  The pivot columns of $A$ tell us the rank of the matrix.  The columns of $A$ without pivots are associated with the free variables, each of which contributes to a basis vector for $N(A)$.  If there are $n$ total columns of $A$, and $r$ of them have pivots, we know that $n-r$ of them don't.  This tells us that the dimension of $N(A)$ is $n-r$. 

Similarly, $A^T$ has $m$ total columns.  If $r$ of them have pivots, then $m-r$ of them don't, which tells us that the dimension of $N(A^T)$ is $m-r$.

These facts about the dimensions of the four fundamental subspaces are summarized in Part 1 of the **Fundamental Theorem of Linear Algebra**. 

**Fundamental Theorem of Linear Algebra (Part 1)**: Let $A$ be $m \times n$.  Then 

<br>

|Space $\quad\quad\quad$|Dimension $\quad\quad\quad\quad\quad$|
|----------------|-----------------------|
|$C(A) \subseteq \R{m}$ | $\textrm{dim}\myb{C(A)}=r$ |
|$N(A) \subseteq \R{n}$ | $\textrm{dim}\myb{N(A)}=n-r$ |
|$C(A^T) \subseteq \R{n}$ | $\textrm{dim}\myb{C(A^T)}=r$ |
|$N(A^T) \subseteq \R{m}$ | $\textrm{dim}\myb{N(A^T)}=m-r$ |



<br>

**Example 2**: Find bases for the four fundamental subspaces of the matrix 

<br>

$$
A = \mymat{
1 & 2 & -2 & 1 \\
-1 & 0 & 0 & 1 \\
0 & 2 & -2 & 2
}
$$

<br>

To find bases for $C(A)$, $N(A)$, and $C(A^T)$ we reduce $A$ to row reduced echelon form. 

<br>

$$
A = \mymat{
1 & 2 & -2 & 1 \\
-1 & 0 & 0 & 1 \\
0 & 2 & -2 & 2
}
\hs \sim \hs
\mymat{
1 & 2 & -2 & 1 \\
0 & 2 & -2 & 2 \\
0 & 2 & -2 & 2
}
\hs \sim \hs
\mymat{
1 & 2 & -2 & 1 \\
0 & 2 & -2 & 2 \\
0 & 0 & 0 & 0
}
$$

<br>

$$
\sim \hs
\mymat{
1 & 2 & -2 & 1 \\
0 & 1 & -1 & 1 \\
0 & 0 & 0 & 0
}
\hs \sim \hs
\mymat{
\color{red}{1} & 0 & 0 & -1 \\
0 & \color{red}{1} & -1 & 1 \\
0 & 0 & 0 & 0
} = R 
$$

<br>

Pivots in the first two columns of $R$ indicate that a basis for $C(A)$ is comprised of the first two columns of $A$.  So we have 

<br>

$$
C(A) = \textrm{span}\mys{\hs\mymat{1 \\ -1 \\ 0}, \mymat{2 \\ 0 \\ 2}\hs}
$$

<br>

Pivots in the first two rows of $R$ indicate that a basis for $C(A^T$ is made up of the first two rows of $R$.  So we have 

<br>

$$
C(A^T) = \textrm{span}\mys{\hs\mymat{1 \\ 0 \\ 0 \\ -1}, \mymat{0 \\ 1 \\ -1 \\ 1}\hs}
$$

<br>

To find a basis for $N(A)$ we write the basic variables $x_1$ and $x_2$ in terms of the free variables $x_3$ and $x_4$.  This gives 

<br>

$$
\begin{array}{rcrr}
x_1 &=& & x_4 \\
x_2 &=& x_3 & -x_4 \\
x_3 &=& x_3 & \\
x_4 &=& & x_4
\end{array}
\quad \Rightarrow \quad 
N(A) = \textrm{span}\mys{\hs\mymat{0 \\ 1 \\ 1 \\ 0}, \mymat{1 \\ -1 \\0 \\ 1}\hs}
$$

<br>

To find a basis for the left nullspace of $A$ we perform row reduction on $A^T$: 

<br>

$$
A^T = 
\mymat{
1 & -1 & 0 \\
2 & 0 & 2 \\
-2 & 0 & -2 \\
1 & 1 & 2
}
\hs\sim\hs
\mymat{
1 & -1 & 0 \\
0 & 2 & 2 \\
0 & -2 & -2 \\
0 & 2 & 2
}
\hs\sim\hs
\mymat{
1 & -1 & 0 \\
0 & 2 & 2 \\
0 & 0 & 0 \\
0 & 0 & 0
}
\hs\sim\hs
\mymat{
1 & 0 & 1 \\
0 & 1 & 1 \\
0 & 0 & 0 \\
0 & 0 & 0
}
$$

<br>

Writing the basic variables in terms of the free variables we have 

<br>

$$
\begin{array}{rcr}
y_1 &=& -y_3 \\
y_2 &=& -y_3 \\
y_3 &=& y_3 
\end{array}
\quad \Rightarrow \quad
N(A^T) = \textrm{span}\mys{\hs\mymat{ -1 \\ -1 \\ 1}\hs}
$$

<br>

Let's check the dimensions of the spaces by counting the number of vectors in each basis.  Remember that the matrix $A$ was $3 \times 4$ so we have $m = 3$ and $n = 4$.  Furthermore, the reduced matrix $R$ had $2$ pivots so the rank of $A$ and $A^T$ is $r = 2$.  We have 

<br>

$$
\begin{array}{rcl}
\dim\myb{C(A)} &=& 2 = r \quad \checkmark\\
& & \\
\dim\myb{N(A)} &=& 2 = 4-2 = n-4 \quad \checkmark\\
& & \\
\dim\myb{C(A^T)} &=& 2 = r \quad \checkmark\\
& & \\
\dim\myb{N(A^T)} &=& 1 = 3-2 = m-4 \quad \checkmark
\end{array}
$$

<br>

Let's group the subspaces and their bases by *where they live*.  The two subspaces of $\R{n} = \R{4}$ were the row space and the nullspace.  Together their bases were 

<br>

$$
\mys{\hs\mymat{1 \\ 0 \\ 0 \\ -1}, \mymat{0 \\ 1 \\ -1 \\ 1}\hs} \quad + \quad \mys{\hs\mymat{0 \\ 1 \\ 1 \\ 0}, \mymat{1 \\ -1 \\0 \\ 1}\hs}
$$

<br>

Note that the dimension of $\R{4}$ is $4$, and combining the bases of $C(A^T)$ and $N(A)$ gave us $4$ vectors.  If you were to check ($\hs$*how would you check?*$\hs$) you'd find that these four vectors are linearly independent. Since we have $4$ linearly independent vectors that live in $\R{4}$, we know that these vectors, taken together, form a basis for the entire space $\R{4}$. 

Similarly, the two subspaces of $\R{m} = \R{3}$ were the column space and the left nullspace.  Together their bases were 

<br>

$$
\mys{\hs\mymat{1 \\ -1 \\ 0}, \mymat{2 \\ 0 \\ 2}\hs} 
\quad + \quad 
\mys{\hs\mymat{1 \\ -1 \\ 1}\hs}
$$

<br>

Here we note that the dimension of $\R{3}$ is 3, and combining the bases of $C(A)$ and $N(A^T)$ gave us $3$ vectors.  Again, if we checked we'd find that these three vectors are linearly independent.  Since we have $3$ lienarly independent vectors that live in $\R{3}$, we know that these vectors, taken together, form a basis for the entire space of $\R{3}$.

Take a second to let this one sink in.  The four **fundamental** subspaces of an $m \times n$ matrix provide bases for each of the larger vector spaces associated with the matrix, namely $\R{m}$ and $\R{n}$.  

**Any** vector in $\R{n}$ can be written as a linear combination of something in the row space of $A$ and the nullspace of $A$. 

**Any** vector in $\R{m}$ can be written as a linear combination of something in the column space of $A$ and the left nullspace of $A$. 

**BUT WE CAN SAY MORE!**

**Any** vector in $N(A)$ is orthogonal to any vector in $C(A^T)$.

**Any** vector in $N(A^T)$ is orthogonal to any vector in $C(A)$.

**Reminder**: We said in class that two vectors $\bv$ and $\bw$ are **orthogonal** if their dot product is zero.  

Let's check that our claims about orthogonality of the fundamental subspaces checks out for the subspaces in Example 1.  

**Example 3**: 

We had $C(A) = \textrm{span}\mys{\hs\mymat{1 \\ 2 \\ 4}\hs}\hs$ and $N(A^T) = \textrm{span}\mys{\hs\mymat{-2 \\ 1 \\ 0}, \mymat{-4 \\ 0 \\ 1}\hs}$

<br>

If we check orthogonality of the basis vectors by taking their dot products, we have 

<br>

$$
\mymat{1 \\ 2 \\ 4} \cdot \mymat{-2 \\ 1 \\ 0} = -2 + 2 = 0
\quad \textrm{and} \quad
\mymat{1 \\ 2 \\ 4} \cdot \mymat{-4 \\ 0 \\ 1} = -4 + 4 = 0 \quad \checkmark
$$

<br>

We also had $C(A^T) = \textrm{span}\mys{\hs\mymat{1 \\ 2}\hs}\hs$ and $N(A) = \textrm{span}\mys{\hs\mymat{-2 \\ 1}\hs}$

<br>

If we check orthogonality of the basis vectors by taking their dot products, we have 

<br>

$$
\mymat{1 \\ 2} \cdot \mymat{-2 \\ 1} = -2 + 2 = 0
\quad \checkmark
$$

<br>

The facts that $C(A)$ and $N(A^T)$ are orthogonal and give us all of $\R{m}$, and $C(A^T)$ and $N(A)$ are orthogonal and give us all of $\R{n}$ is summarized in Part 2 of the Fundamental Theorem of Linear Algebra

**Fundamental Theorem of Linear Algebra (Part 2)**: 

* $C(A)$ and $N(A^T)$ are orthogonal subspaces in $\R{m}$
* $C(A^T)$ and $N(A)$ are orthogonal subspaces in $\R{n}$

OK, hold on a sec.  In the previous example we showed that the basis vectors of each matched pair of fundamental subspaces were orthogonal.  Is that enough to show that **any** pair of vectors from the matched subspaces are orthogonal?  

It turns out that it is.  To see this let $S_1 = \textrm{span}\mys{\hs \bv_1, \bv_2, \ldots, \bv_k}$ and $S_2 = \textrm{span}\mys{\hs\bw_1, \bw_2, \ldots, \bw_\ell}$ be two subspaces in $\R{n}$ and suppose that all of the $\bv$'s are orthogonal to all of the $\bw$'s.  We want to show that any arbitrary vector in $S_1$ is orthogonal to any arbitrary vector in $S_2$. 

Let $\bx = \alpha_1\bv_1 + \alpha_2 \bv_2 + \cdots + \alpha_k\bv_k$ be an arbitrary vector in $S_1$, and $\by = \beta_1\bw_1 + \beta_2 \bw_2 + \cdots + \beta_\ell\bw_\ell$ be an arbitrary vector in $S_2$.  If we dot $\bx$ and $\by$ together we get 

<br>

$$
\bx \cdot \by = \myp{\alpha_1\bv_1 + \alpha_2 \bv_2 + \cdots + \alpha_k\bv_k}\cdot\myp{\beta_1\bw_1 + \beta_2 \bw_2 + \cdots + \beta_\ell\bw_\ell}
$$

<br>

Multiplying and distributing the sums we have 

<br>

$$
\begin{array}{rcl}
\bx \cdot \by &=& \alpha_1\beta_1\myp{\bv_1 \cdot \bw_1} + 
\alpha_1\beta_2\myp{\bv_1 \cdot \bw_2} + \cdots + 
\alpha_1\beta_\ell\myp{\bv_1 \cdot \bw_\ell} + \\
&& \alpha_2\beta_1\myp{\bv_2 \cdot \bw_1} + 
\alpha_2\beta_2\myp{\bv_2 \cdot \bw_2} + \cdots + 
\alpha_2\beta_\ell\myp{\bv_2 \cdot \bw_\ell} + \\
&& ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ \vdots\\
&&\alpha_k\beta_1\myp{\bv_k \cdot \bw_1} + 
\alpha_k\beta_2\myp{\bv_k \cdot \bw_2} + \cdots + 
\alpha_k\beta_\ell\myp{\bv_k \cdot \bw_\ell}  
\end{array}
$$

<br>

Now note that since the basis vectors are orthogonal, each of the dot products in the gigantic sum above are zero, so we have 

<br>

$$
\bx \cdot \by = 0 
$$

<br>

OK, so we've shown for a particular example that $C(A)$ is orthogonal to $N(A^T)$ and $C(A^T)$ is orthogonal to $N(A)$.  Now it's time to show it in general.  First we have to find a simple mathy way of writing an arbitrary element in each of the four fundamental subspaces.

Note that a vector in $C(A)$ is an arbitrary linear combination of the columns of $A$.  One way to write a linear combination of the columns of $A$ is to multiply $A$ by a vector of coefficients $\bx$.  Similarly, we can write a linear combination of the columns of $A^T$ by multiplying $A^T$ by some vector $\by$.  We then have 

* $C(A)$ = All vectors $A\bx$ for some $\bx$ in $\R{n}$
* $N(A^T)$ = All vectors $\by$ in $\R{n}$ such that $A^T\by = {\bf 0}$

   

* $C(A^T)$ = All vectors $A^T\by$ for some $\by$ in $\R{m}$
* $N(A)$ = All vectors $\bx$ in $\R{m}$ such that $A\bx = {\bf 0}$

Further, we need to recall that another way to write $\bv \cdot \bw$ is $\bv^T\bw$. 

To show that $C(A)$ is orthgonal to $N(A^T)$ we dot an arbitrary vector of the form $A\bx$ by a vector $\by$ such that $A^T\by = {\bf 0}$.  We have 

<br>

$$
(A\bx)\cdot\by = (A\bx)^T\by = \bx^TA^T\by = \bx^T{\bf 0} = 0 
$$

<br>

To show that $C(A^T)$ is orthgonal to $N(A)$ we dot an arbitrary vector of the form $A^T\by$ by a vector $\bx$ such that $A\bx = {\bf 0}$.  We have 

<br>

$$
(A^T\by)\cdot\bx = (A^T\by)^T\bx = \by^TA\bx = \by^T{\bf 0} = 0 
$$
