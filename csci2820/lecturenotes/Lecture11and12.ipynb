Let's revisit the first example from the previous lecture, but ask some different questions.  

**Example 1**: Again consider $A = \mymat{1 & 1 & 2 & 4 \\ 1 & 2 & 2 & 5 \\ 1 & 3 & 2 & 6}$

**Questions**: 

1. Which are the important columns of $A$? 
2. What is the rank of $A$? 
3. What are the special solutions of $A\bx = {\bf 0}$? 
4. What is the complete solution of $A\bx = {\bf b}$ for ${\bf b} = \mymat{0, & -1, & -2}^T$

<br>

 1.$~~$Row reducing $A$ to reduced row echelon form we have 

<br>

$$
A = 
\mymat{
1 & 1 & 2 & 4 \\ 
1 & 2 & 2 & 5 \\ 
1 & 3 & 2 & 6
}
\quad \sim \quad 
\mymat{
1 & 0 & 2 & 3 \\ 
0 & 1 & 0 & 1 \\ 
0 & 0 & 0 & 0
}
= R 
$$

<br>

We argued last time that the important columns of $A$ were columns 1 and 2, because those are the columns with pivots.  Columns 3 and 4 were redundant because they could be written as linear combinations of columns 1 and 2 (col 3 is 2 times col 1 and col 4 is 3 times col 1 plus column 2).  In other words, the vectors in columns 1 and 2 are the vectors that contribute unique information to the column space of $A$. 

<br> 

2.$~~$The rank of $A$ is $r = 2$ because there are two pivots in the reduced system. Notice that the rank also corresponds to the number of important columns in $A$. 

<br> 

3.$~~$The special solutions of $A\bx = {\bf 0}$ are obtained by writing the basic variables $x_1$ and $x_2$ in terms of the free variables $x_3$ and $x_4$: 

<br>

$$
\begin{array}{lcrr}
x_1 &=& -2x_3 & -3x_4 \\
x_2 &=& & -x_4 \\
x_3 &=& x_3 & \\
x_4 &=& & x_4
\end{array}
\quad \Rightarrow \quad 
\bx = x_3 \mymat{-2 \\ 0 \\ 1 \\ 0} 
+ x_4 \mymat{-3 \\ -1 \\ 0 \\ 1}
$$

<br> 

Thus the special solutions are 

<br> 

$$
{\bf s}_1 =  \mymat{-2 \\ 0 \\ 1 \\ 0}
\quad \quad \textrm{and} \quad \quad 
{\bf s}_2 = \mymat{-3 \\ -1 \\ 0 \\ 1}
$$

<br> 

Remember that the special solutions are in the Nullspace of $A$, as is any linear combination of those special solutions. 

<br>

4.$~~$To find the complete solution of $A\bx = {\bf b}$ for a specific rhs vector we form the augmented system and do row reduction.  For this $A$ and ${\bf b}$ the reduced augmented system is 

<br>

$$
\left[
\begin{array}{rrrr|r}
1 & 1 & 2 & 4 & 0  \\ 
1 & 2 & 2 & 5 & -1\\ 
1 & 3 & 2 & 6 & 2
\end{array}
\right]
\quad \sim \quad 
\left[
\begin{array}{rrrr|r}
1 & 0 & 2 & 3 & 1 \\ 
0 & 1 & 0 & 1 & -1 \\ 
0 & 0 & 0 & 0 & 0
\end{array}
\right]
$$

<br>

Solving for the basic variables in terms of the free variables **and** the rhs vector gives 

<br> 

$$
\begin{array}{lcrrr}
x_1 &=& 1 & -2x_3 & -3x_4 \\
x_2 &=& -1 & & -x_4 \\
x_3 &=& & x_3 & \\
x_4 &=& & & x_4
\end{array}
\quad \Rightarrow \quad 
\bx = \mymat{1 \\ -1 \\ 0 \\ 0} +  
x_3 \mymat{-2 \\ 0 \\ 1 \\ 0} 
+ x_4 \mymat{-3 \\ -1 \\ 0 \\ 1}
$$

<br>

Remember that the vector without the free variables is the **particular** solution of $A\bx = {\bf b}$ which gaurantees that multiplication of $\bx$ by $A$ gives the appropriate rhs vector ${\bf b}$.  The vectors corresponding to the free variables are again the special solutions, and don't mess up the rhs result because multiplication of these vectors by $A$ gives the zero vector. 

<br> 

OK, one more question:  Is $A\bx = {\bf b}$ solvable for all ${\bf b}$ in $\mathbb{R}^3$? 

Clearly not, because the reduced row echelon form of $A$ did not have a pivot in each row.  Had row reduction left a nonzero value in the third entry in the rhs vector, the last row would tell us that the system had no solution. 

So what went wrong?  The problem is that even though $A$ had four columns, two of them were redundant.  When some vectors in a set are redundant we call the set of vectors **linearly dependent**. 

When no vectors in a set are redundant we say the set is **linearly independent**. 

<br>

**Examples in ** $\mathbb{R}^2$: 

<br>

Let $\bv_1 = \mymat{1 \\ 0}$ and $\bv_2 = \mymat{0 \\ 1}$.  Clearly we can't write $\bv_2$ as a linear combination of $\bv_1$, and visa versa, so together they form a linearly independent set. 

Furthermore, we can get anywhere in $\mathbb{R}^2$ by taking linear combinations of $\bv_1$ and $\bv_2$, so we say $\bv_1$ and $\bv_2$ **span** $\mathbb{R}^2$ and write $\textrm{span}\left\{ \bv_1, \bv_2 \right\} = \mathbb{R}^2$. 

<br>

Now consider $\bv_1 = \mymat{1 \\ 0}$, $\bv_2 = \mymat{0 \\ 1}$, $\bv_3 = \mymat{2 \\ 3}$.  Since we can write $\bv_3 = 3\bv_1 + 2\bv_2$ the set is linearly dependent.  But, we can still say the vectors span $\mathbb{R}^2$ and write $\textrm{span}\left\{ \bv_1, \bv_2, \bv_3 \right\} = \mathbb{R}^2$ since any vector in $\mathbb{R}^2$ can be written as a linear combination of $\bv_1$, $\bv_2$, and $\bv_3$ (for instance, by completely ignoring $\bv_3$). 

<br> 

Finally, consider $\bw_1 =\mymat{1 \\ 2}$ and $\bw_2 = \mymat{-2 \\ -4}$.  Clearly this set is linearly dependent, since $\bw_2 = -2\bw_1$. But this time the set of vectors do **not** span $\mathbb{R}^2$, since taking linearly combinations of $\bw_1$ and $\bw_2$ can only result in a scalar multiple of $\bw_1$.  In other words, the span of $\bw_1$ and $\bw_2$ is the line containing both of the vectors. 

<br>

The formal mathematical definition of linear independence is as follows: 

<br>

**Fact**: A set $\bv_1, \bv_2, \ldots, \bv_n$ is linearly dependent if and only if the only linear combination that produces the zero vector ${\bf 0}$ is trivial linear combination.  In other words, if the set is linearly independent, and 

<br>

$$
c_1\bv_1 + c_2\bv_2 + \cdots + c_n \bv_n = {\bf 0}
$$

<br>

then it must be the case that $c_1 = c_2 = \cdots = c_n = 0$. 

<br>

This particular definition leads us to a simple test for linear independence.  If we build a matrix with the set of vectors as its columns

<br>

$$
A = \mymat{\bv_1 & \bv_2 & \cdots & \bv_n}
$$

<br>

then the set is linearly independent if and only if the system $A{\bf c} = {\bf 0}$ has only the zero solution. 

<br>

**Example**  Let's check the second example from above.  Putting vectors $\bv_1$, $\bv_2$, and $\bv_3$ in a matrix, we have 

<br>

$$
A = \mymat{1 & 0 & 2 \\ 0 & 1 & 3} 
$$

<br>

Usually the next step would be to perform row reduction on $A$ to put it in reduced row echelon form, but for this particular example, $A$ is already in its reduced form.  Notice that $A$ does not have a pivot in the third column.  This means that $A$ has one free variable, and thus the system $A{\bf c} = {\bf 0}$ has infinitely many solutions, each of which prescribes a particular combination of the vectors $\bv_1$, $\bv_2$, $\bv_3$ that would give the zero vector.  This means that the set is linearly dependent. 

<br>

So what about the question of whether or not the vectors span $\mathbb{R}^2$?  It's clear from the reduced matrix that they do in fact, since there is a pivot in every row.  

<br>

**Summary**:  If you put a set of a vectors in a matrix, row reduce, and find a column without a pivot, the set of vectors is linearly dependent.  If each column has a pivot then the set of vectors is linearly independent. If there is a pivot in every row of the reduced matrix, then the set of vectors spans the space they live in.  If there is not a pivot in every row the vectors do not span the space they live in. 

<br>

**Example**: Let's go back to the matrix that we started the day with.  Putting it in reduced row echelon form we have. 

<br>

$$
A = \mymat{
1 & 1 & 2 & 4 \\ 
1 & 2 & 2 & 5 \\ 
1 & 3 & 2 & 6
}
\quad \sim \quad 
\mymat{
1 & 0 & 2 & 3 \\ 
0 & 1 & 0 & 1 \\ 
0 & 0 & 0 & 0
} = R 
$$

<br>

Since the reduced matrix does not have a pivot in every column, the columns of $A$ are not linearly independent.  Furthermore, since the reduced matrix does not have a pivot in every row, the column space of $A$ is not all of $\mathbb{R}^3$. 

**Question**: Suppose you have $n$ vectors that live in $\mathbb{R}^m$.  If $n > m$ can the set of vectors every be linearly independent? 

<br>

**Answer**: Nope!  Put the vectors in a matrix and row reduce.  Since the matrix has more columns than rows its reduced system is guaranteed to have at least one free variable, so the columns are not linearly independent. 

<br>

**Question**: Suppose you have $n$ vectors that live in $\mathbb{R}^m$.  If $n < m$ can the set of vectors ever span $\mathbb{R}^m$? 

<br>

**Answer**: Nope! Since the matrix has more rows than columns its reduced system is guaranteed to have at least one row that does not have a pivot, so the vectors cannot span $\mathbb{R}^m$. 

<br>

OK, now we're ready to use what we've learned to prove a big result about nonsingular matrices

**Fact**: A square $n \times n$ matrix $A$ is nonsingular if and only if its columns are linearly independent. 


**Proof**: Recall that $A$ nonsingular is equivalent to the statement that $A\bx = {\bf b}$ has a unique solution for each ${\bf b}$ in $\mathbb{R}^n$. It is this second characterization of nonsingularity that we will use.

We know that if $A$ has $n$ linearly independent columns, then row reduction will produce a pivot in each column.  Since the matrix is square this implies that there is also a pivot in each row.  

Having a pivot in each row implies that the columns of $A$ span all of $\mathbb{R}^n$, which means that $A\bx = {\bf b}$ has a solution for all ${\bf b}$ in $\mathbb{R}^n$. 

Having a pivot in each column implies that the system has no free variables.  Since there are no free variables the solution to $A\bx = {\bf b}$ is unique.  

Thus $A$ is nonsingular.

OK, back to our favorite example again 

<br>

$$
A = \mymat{
1 & 1 & 2 & 4 \\ 
1 & 2 & 2 & 5 \\ 
1 & 3 & 2 & 6
}
\quad \sim \quad 
\mymat{
1 & 0 & 2 & 3 \\ 
0 & 1 & 0 & 1 \\ 
0 & 0 & 0 & 0
} = R 
$$

<br>

The important vectors for the column space of $A$ were $\bv_1 = \mymat{1 \\ 1 \\ 1}$ and $\bv_2 = \mymat{1 \\ 2 \\ 3}$.  

<br>

You can get anywhere in $C(A)$ by taking linear combinations of $\bv_1$ and $\bv_2$.  Note that adding $\bv_3$ or $\bv_4$ doesn't help you get to new places, because $\bv_3$ and $\bv_4$ can themselves be written as linear combinations of $\bv_1$ and $\bv_2$. 

So including $\bv_3$ and $\bv_4$ in our description of the column space does not help, because they are redundant. What if we removed either $\bv_1$ or $\bv_2$?  Would we still be able to get everywhere in $C(A)$?  

**Nope!**  So adding vectors beyond $\bv_1$ and $\bv_2$ is redundant, but removing either of them would cripple our ability to get places in $C(A)$.  When we have a set such as this, we call it a **basis** for the space that they describe. 

<br>

**Definition**: A **basis** for a space is a set of vectors that: 
1. are linearly independent
2. span the entire space 

<br>

**Question**: What is a basis for $\mathbb{R}^2$? 

**Answer**: We need to find a set of linearly independent vectors that span all of $\mathbb{R}^2$.  One obvious example comes to mind, that is the set of the standard basis vectors in $\mathbb{R}^2$: 

<br>

$$
\be_1 = \mymat{1 \\ 0} \quad \be_2 = \mymat{0 \\ 1}
$$

<br>

**Question**: Is that the only basis for $\mathbb{R}^2$? 

**Answer**: Nope!  There are lots (like, infinitely many).  We just need to find any set of linearly independent vectors that span $\mathbb{R}^2$.  Another example is 

<br>

$$
\bv_1 = \mymat{1 \\ 1} \quad \bv_2 = \mymat{1 \\ -1}
$$

<br>

**Definition**: The **dimension** of a space is equal to the number of elements in any basis for the space.  

<br>

**Example**: The dimension of $\mathbb{R}^2$ is 2.  Sometimes we write $\textrm{dim}( \mathbb{R}^2 ) = 2$. 

<br>

**Example**: One more time consider the row reduced $A$: 

<br>

$$
A = \mymat{
1 & 1 & 2 & 4 \\ 
1 & 2 & 2 & 5 \\ 
1 & 3 & 2 & 6
}
\quad \sim \quad 
\mymat{
1 & 0 & 2 & 3 \\ 
0 & 1 & 0 & 1 \\ 
0 & 0 & 0 & 0
} = R 
$$

<br> 

A basis for the column space of $A$ are the first two columns $A$.  So we can write 

<br>

$$
C(A) = \textrm{span}\left\{~~ \mymat{1 \\ 1 \\ 1}, \mymat{1 \\ 2 \\ 3}~~\right\}
$$

<br>

Note that the basis vectors for the column space correspond to the columns of the original $A$ corresponding to the pivot columns or $R$.  The dimension of $C(A)$ for this particular $A$ is 2, since its basis contains two vectors.  Note also that this is exactly the **rank** of $A$. 

<br> 

**Question**: What is a basis for the nullspace of $A$? 

**Answer**: Recall that the nullspace of $A$ is all linear combinations of the special solutions of $A\bx = {\bf 0}$.  So the special solutions form a basis for $N(A)$.  

<br>

For this particular example the special solutions were 

<br>

$$
{\bf s}_1 = \mymat{-2 \\ 0 \\ 1 \\ 0} \quad {\bf s}_2 = \mymat{-3 \\ -1 \\ 0 \\ 1}
$$

<br>

So we can write $N(A) = \textrm{span}\left\{ ~~ \mymat{-2 \\ 0 \\ 1 \\ 0}, \mymat{-3 \\ -1 \\ 0 \\ 1} ~~ \right\}$

Note that in this case the dimension of the nullspace of $A$ is also 2.

###Matrix Spaces

We can also find bases for spaces of matrices.  For now we'll just consider subspaces of $\mathbb{M}_{2 \times 2}$. 

<br>

**Example**: Let $S$ be the space of all upper triangular $2\times 2$ matrices.  

To find a basis for $S$ we want to find a set of linearly independent basis elements that span the set.  In other words, we want to find a non-redundant set of $2 \times 2$ matrices such that any upper triangular matrix can be written as a linear combination of the set. 

Consider an arbitrary upper triangular $2 \times 2$ matrix: $\mymat{a & b \\ 0 & d}$ and note that it can be written as  

<br>

$$
\mymat{a & b \\ 0 & d } = 
a\mymat{1 & 0 \\ 0 & 0} + 
b\mymat{0 & 1 \\ 0 & 0} + 
c\mymat{0 & 0 \\ 0 & 1}  
$$

<br>

This shows that the three matrices above span the space of all upper triangular $2 \times 2$ matrices.  For the set to be a basis it must also be the case that they are linearly indendent.  It's easy to see that the only linear combination of the three matrices that results in the zero matrix is the trivial combination, so the set is linearly independent. 

Since the basis for the space has three elements, the dimension of the space is $3$.
