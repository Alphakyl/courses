So far we've talked about the solution to $A\bx = {\bf b}$ in a couple of different ways.  

* Sol exists if ${\bf b}$ is a linear combination of the columns of $A$
* Sol exists if $A$ is square and nonsingular
* Sol exists if $A$ is square and elimination produces a pivot in ever row

In this chapter we start to look at the solvability of $A\bx = {\bf b}$ from a much richer perspective, and we also relax the assumption that $A$ must be square. In order to do this we must learn about the new concepts of vector spaces and subspaces. 

###Vector Spaces

Essentially a vector space is a set of vectors that satisfy some very specific properties and axioms.  In general, arithmetic in a vector space must satisfy the usual arithmetic properties: commutativity of addition, associativity of addition, distribution laws, etc. 

In addition to these things there are two very important rules that a collection of vectors must satisfy in order to be a vector space: 

1. **closure under vector addition**: If vectors ${\bf u}$ and ${\bf v}$ are in the space, then so is ${\bf u} + {\bf v}$
2. **closure under scalar multiplcation**: If ${\bf u}$ is in the space then so is $c{\bf u}$

    Notice that closure under scalar multiplication automatically assumes that the zero vector is in the space (since we could multiply a vector by 0), but this is such a useful rule on it's own for checking that a set of vectors is a vector space that we give it honorary rule status $~~~~~~~~~~~$
    
3. **contains the zero vector**: The zero vector ${\bf 0}$ is in the space

If you prefer brevity instead of verbosity then both of the closure rules can be combined into one by saying that any linear combinations of two vectors are in the space. 

1. **closure under linear combination**: If vectors ${\bf u}$ and $\bv$ are in the space then so is $c\bu + d\bv$. 

We've already seen several vector spaces this semester.  The most important one for us is the space of all $n$-length vectors with real values.   

**Definition**: The space of all real vectors of length $n$ is $\mathbb{R}^n$

We know that $\mathbb{R}^n$ satisfies all of the usual rules of vector arithmetic.  To see that it satisfies the closure rules we note that: 

1. If you add two $n$-length vectors you get an $n$-length vector back 
2. If you multiply an $n$-length vector by a scalar you get an $n$-length vector back.  

Another important vector space is the space of all $n$-length complex-valued vector spaces: $\mathbb{C}^n$.

Other things can be vector spaces too. 

**Definition**: $\mathbb{M}_{2\times 2}$ is the space of all real $2 \times 2$ matrices. 

Note that if you add two $2 \times 2$ matrices together, or multiply a $2 \times 2$ matrix by a scalar, you get back a $2 \times 2$ matrix.  Furthermore the zero "vector" in this space is the $2 \times 2$ matrix of all zeros. 

OK, so vector spaces are important.  They're the setting in which all linear algebra takes place.  But even more interesting are what are called **subspaces** of a vector space.  A subspace is a vector space *inside* another vector space. 

**Definition**: A subspace of a vector space is a set of vectors (including ${\bf 0}$) that satisfies the rules of being a vector space.  Let $S$ be a set inside of a vector space, then if $\bu$ and $\bv$ are in $S$

1. $\bu + \bv$ is in $S$
2. $c\bu$ is in $S$

**Example 1**:  Let $S$ be the set of all vectors in $\mathbb{R}^3$ of the form $~~\mymat{x \\ y \\ 0}$

It's clear that if we let $x = y = 0$ then we get the zero vector, so the zero vector is in $S$

If we take two vectors in the space and add them together we get 

<br>

$$
\mymat{
x_1 \\ y_1 \\ 0
}
\mymat{
x_2 \\ y_2 \\ 0
}
=
\mymat{
x_1+x_2 \\ y_1+y_2 \\ 0
}
$$

<br> 

Clearly this sum is in $S$ since it's only nonzero entries are in positions 1 and 2. 

Finally, if we multiply a vector in $S$ by a scalar $c$ we get 

<br>

$$
c\mymat{x \\ y \\ 0} = \mymat{cx \\ cy \\ 0} 
$$

<br>

which is in $S$. Having shown that $S$ satisfies the closure properties we can conclude that $S$ is a subspace of $\mathbb{R}^3$. 

**Note**:  This subspace is all vectors that lie in the $xy$-plane, but it is **NOT** the same as $\mathbb{R}^2$ because the vectors in $S$ all have length 3. 

**Example 2**: Let $S$ be the set of all vectors in $\mathbb{R}^3$ that are multiples of $\mymat{1 \\ 1 \\ 1}$

Clearly the zero vector is in $S$ because we could multiply the ones vector by zero.  If we sum two vectors in $S$ we get, e.g. 

<br>

$$
\mymat{c \\ c \\ c} + \mymat{ d \\ d \\ d} = \mymat{c+d \\ c+d \\ c+d} = \myp{c+d}\mymat{1 \\ 1 \\1 }
$$

<br>

which shows that $S$ is closed under addition.  Showing that $S$ is closed under scalar multiplication is trivial: 

<br>

$$
d\mymat{c \\ c \\ c} = \myp{dc}\mymat{1 \\ 1 \\ 1}
$$

<br>

Thus we've shown that all multiples of the length-3 ones vector is a subspace of $\mathbb{R}^3$.  If we think about the space geometrically we realize that $S$ is all vectors that lie on the line through the origin that makes a 45 degree angle with each of the coordinate axes.  

In general all interesting subspaces of $\mathbb{R}^3$ are either lines that go through the origin, or planes that contain the origin.  More trivial subspaces of $\mathbb{R}^3$ are the so-called *trivial subspace* which just contains the zero vector, and the entirity of $\mathbb{R}^3$ (in other words, any vector space is a subspace of itself). 

Here are a couple examples of subspaces of $\mathbb{M}_{2 \times 2}$ that you should try to prove yourself. 

**Example 3**: The set of all real $2 \times 2$ upper triangular matrices. 

**Example 4**: The set of all real $2 \times 2$ diagonal matrices. 

OK, how about some examples of things that **aren't** subspaces. 

**Example 5**: The set of all vectors $\mymat{x \\ y \\ z}$ such that $x + y + z + 1 = 0$

This space fails the closure requirments in several different ways.  The most obvious is the fact that the zero vector is not included in the set since 

<br>

$$
0 + 0 + 0 + 1 \neq 0 
$$

<br>

Alternatively, note that $\mymat{-1 \\ 0 \\ 0}$ and $\mymat{1 \\ -2 \\ 0}$ are in the space, but 

<br>

$$
\mymat{-1 \\ 0 \\ 0} + \mymat{1 \\ -2 \\ 0} = \mymat{0 \\ -2 \\ 0} ~~~\textrm{but}~~~~ 0 - 2 + 0 + 1 \neq 0 
$$

<br>

Similarly 

$$
3\mymat{-1 \\ 0 \\ 0} = \mymat{-3 \\ 0 \\ 0} ~~~\textrm{but}~~~~ -3 +0 + 0 + 1 \neq 0 
$$

Note that in order to prove that a set is **not** a subspace you really only need to show that it fails **one** of the conditions. 

**Example 6**: The set of all vectors $\mymat{x \\ y}$ in $\mathbb{R}^2$ s.t. $x \geq 0$ and $y \geq 0$

This set does contain the zero vector, and actually satisfies closure under addition.  But it fails closure under scalar multiplcation, since 

<br>

$$
-1\mymat{1 \\ 1} = \mymat{-1 \\ -1} \quad \textrm{which isn't in the space}
$$

<br>

One **extremely** important subspace of $\mathbb{R}^n$ is the *span* of a fixed number of vectors. 

**Definition**: Let $\bu_1, \bu_2, \ldots, \bu_k$ be fixed vectors.  Then $\textrm{span}\left\{ \bu_1, \bu_2, \ldots, \bu_k\right\}$ is the set of all linear combinations of $\bu_1, \bu_2, \ldots, \bu_k$.

**Example 7**: Let $\bu = \mymat{1 \\ 2 \\ 3}$ and $\bv = \mymat{-2 \\ 0 \\ 1}$.  Then 

<br>

$$
\textrm{span}\left\{\bu,\bv \right\} = ~~~\textrm{all vectors of the form }~~~ c\mymat{1 \\ 2 \\ 3} + d\mymat{-2 \\ 0 \\ 1}
$$

<br>

OK, let's see how subspaces can help us say something about the solvability of $A\bx = {\bf b}$. To do this we need to define a very important subspace associated with a given matrix. 

**Definition**: The **column space** of an $m \times n$ matrix $A$, written as $C\myp{A}$, is the set of all linear combinations of the columns of $A$.  Or in other words, the column space of $A$ is the span of the columns of $A$. 

Why is this important? Recall that one way we interpreted matrix-vector multiplication was as a linear combination of the columns of $A$ where the coefficients of the linear combination are the entries of $\bx$. 

Furthermore, this means that $A\bx = {\bf b}$ has a solution exactly when the rhs vector ${\bf b}$ is a linear combination of the columns of $A$.  Or in other wrods, $A \bx = {\bf b}$ has a solution when ${\bf b}$ is in the column space of $A$. 

**Example 8**: Consider the matrix $A = \mymat{1 & 2 \\ 2 & -1 \\ 3 & 2}$

Where does $C\myp{A}$ live?  Well, any linear combination of the columns of $A$ will have length-3.  Thus for this example $C\myp{A}$ is a subspace of $\mathbb{R}^3$. 

What are some vectors in $C\myp{A}$?  Some trivial ones are 

<br>

$$
\mymat{1 \\ 2 \\ 3} ~~~ \textrm{and} ~~~ \mymat{2 \\ -1 \\ 2}
$$

<br>

Some less trivial ones are 

<br>

$$
\mymat{3 \\ 1 \\ 5} ~~~ \textrm{and} ~~~ \mymat{-1 \\ 3 \\ 1}
$$

<br>

**Example 9**:  Describe the column spaces of the following matrices 

<br>

$$
I = \mymat{1 & 0 \\ 0 & 1} ~~~ \textrm{and} ~~~ A = \mymat{1 & 2 \\ 3 & 6}
$$

<br>

The column space of the $2 \times 2$ identity matrix is the entirity of $\mathbb{R}^2$, since for any vector in $\mathbb{R}^2$ we can write 

<br>

$$
\mymat{x \\ y} = x\mymat{1 \\ 0} + y\mymat{0 \\ 1} = \mymat{1 & 0 \\ 0 & 1}\mymat{x \\ y}
$$

<br>

The column space of $A$ on the other hand is **not** all of $\mathbb{R}^2$.  Note that 

<br>

$$
2\mymat{1 \\ 3} + 3\mymat{2 \\ 6} = 
2\mymat{1 \\ 3} + 6\mymat{1 \\ 3} =
8\mymat{1 \\ 3}   
$$
<br>

Linear combinations of the columns of $A$ only produce scalar multiples of the vector $\mymat{1 & 3}^T$.  This is simply a line in the $xy$-plane. 

###The Nullspace of A

Another super important subspace associated with a matrix $A$ is the set of all vectors $\bx$ such that $A\bx = {\bf 0}$.  This is called the **nullspace** of $A$ and written as $N\myp{A}$.  

Note that for any matrix $A$ there is at least one vector in its nullspace, namely the zero vector ${\bf 0}$.  For nonsingular/invertible matrices the zero vector is the only nullspace vector.  But for singular matrices and some rectangular matrices there are typically many more.

Not surprisingly we find them through elimination. 

**Example 10**: Find the nullspace of $A = \mymat{1 & 3 \\ 2 & 6}$

Performing elimination we have 

<br>

$$
\myaug{
1 & 2 && 0 \\
3 & 6 && 0 
} 
~~ \sim ~~
\myaug{
1 & 2 && 0 \\
0 & 0 && 0 
}
$$

<br>

Thus $x_1$ is a basic variable and $x_2$ is a free variable.  Writing the basic variable in terms of the free variable we have 

<br>

$$
1x_1 + 2x_2 = 0 \quad \Rightarrow \quad 
\begin{eqnarray}
x_1 &=& -2x_2 \\
x_2 &=& x_2 
\end{eqnarray}
\quad \Rightarrow \quad
\bx = t \mymat{-2 \\ 1}
$$

<br>

So we've found that ${\bf s}_1 = \mymat{-2 \\ 1}$ is in $N\myp{A}$ as is any scalar multiple of $\mymat{-2 \\ 1}$. 

Let's check a few: 

<br>

$$
\mymat{1 & 2 \\ 3 & 6}\mymat{-2 \\ 1} = \mymat{-2+2 \\ -6+6} = \mymat{0 \\ 0}
$$
<br>

$$
\mymat{1 & 2 \\ 3 & 6}\mymat{2 \\ -1} = \mymat{2-2 \\ 6-6} = \mymat{0 \\ 0}
$$
<br>

$$
\mymat{1 & 2 \\ 3 & 6}\mymat{-6 \\ 3} = \mymat{-6-6 \\ -18+18} = \mymat{0 \\ 0}
$$

<br>

**Definition**: The nonzero vector ${\bf s}_1$ is called a *special solution* of $A\bx = {\bf 0}$. 

<br> 

**Example 11**: Find all solutions to $A\bx = {\bf 0}$ for $A = \mymat{1 & 2 \\ 2 & 6}$

Performing elimination we have

<br>

$$
\myaug{
1 & 2 && 0 \\
2 & 6 && 0 
}
\quad \sim \quad 
\myaug{
1 & 2 && 0 \\
0 & 2 && 0 
}
$$

<br>

Since the upper triangular matrix $U$ has 2 pivots we know that the only solution to $A\bx = {\bf 0}$ is $\bx = \mymat{0 & 0}^T = {\bf 0}$. The matrix $A$ is nonsingular and thus has no special solutions to $A\bx = {\bf 0}$ and the nullspace of $A$ is just $N\myp{A} = \left\{{\bf 0}\right\}$.

**Example 12**: Find all special solutions to $B\bx = {\bf 0}$ for $B = \mymat{1 & 2 & 2 & 4 \\ 2 & 6 & 4 & 12}$

Since we know that the augmented zero vector is not going to change during elimination, we'll leave it off and just work on the matrix. 

<br>

$$
\mymat{
1 & 2 & 2 & 4 \\
2 & 6 & 4 & 12 
}
\quad \sim \quad
\mymat{
1 & 2 & 2 & 4 \\
0 & 2 & 0 & 4 
} = U
$$

<br>

Notice that the variables $x_1$ and $x_2$ are basic while the variables $x_3$ and $x_4$ are free.  At this point we should solve for the basic variables in terms of the free variables to find the special solutions.  However, things get a little bit easier if we take elimination a few steps further. 

<br>

$$
\sim \quad 
\mymat{
1 & 0 & 2 & 0 \\
0 & 2 & 0 & 4 
}
\quad \sim \quad
\mymat{
1 & 0 & 2 & 0 \\
0 & 1 & 0 & 2
} = R
$$

<br>

We then have 

<br>

$$
\begin{eqnarray}
x_1 &=& -2x_3 \\
x_2 &=& -2x_4 \\
x_3 &=& x_3 \\
x_4 &=& x_4 
\end{eqnarray}
\quad \Rightarrow \quad
\bx = 
x_3
\mymat{
-2 \\ 0 \\ 1 \\ 0
} + 
x_4
\mymat{
0 \\ -2 \\ 0 \\ 1
}
$$

<br>

Note that because we had two free variables we found two separate components of the solution to $B\bx = {\bf b}$.  These separate solutions are the special solutions of $B\bx = {\bf 0}$. 

<br> 

$$
{\bf s}_1 = 
\mymat{
-2 \\ 0 \\ 1 \\ 0
} 
\quad \textrm{and} \quad
{\bf s}_2 = 
\mymat{
0 \\ -2 \\ 0 \\ 1
}
$$

<br> 

We can check that the individual special solutions each solve $B\bx = {\bf 0}$.  

<br>

$$
B{\bf s}_1 = 
\mymat{
1 & 2 & 2 & 4 \\
2 & 6 & 4 & 12 
}
\mymat{
-2 \\ 0 \\ 1 \\ 0
} = 
\mymat{
-2 + 0 + 2 + 0 \\
-4 + 0 + 4 + 0
}
= \mymat{0 \\ 0}
$$

<br>

$$
B{\bf s}_2 = 
\mymat{
1 & 2 & 2 & 4 \\
2 & 6 & 4 & 12 
}
\mymat{
0 \\ -2 \\ 0 \\ 1
} = 
\mymat{
0 + -4 + 0 + 4 \\
0 + -6 + 0 + 12
}
= \mymat{0 \\ 0}
$$

<br>

**BUT WAIT! THERE'S MORE!**

If we take any linear combination of the special solutions we get another solution to $B\bx = {\bf 0}$.  For example, suppose we took ${\bf s} = 2{\bf s}_1 - {\bf s}_2$. Then 

<br>

$$
{\bf s} = \mymat{-4 \\ 2 \\ 2 \\ -1} \quad \textrm{and} \quad 
B{\bf s} = 
\mymat{
1 & 2 & 2 & 4 \\
2 & 6 & 4 & 12 
}
\mymat{
-4 \\ 2 \\ 2 \\ -1
} = 
\mymat{
-4 + 4 + 4 - 4 \\
-8 + 12 + 8 - 12
}
= \mymat{0 \\ 0}
$$

<br>

The special solutions give us an excellent way of characterizing the nullspace of a matrix, that is, the nullspace of $A$ is the **span of its special solutions**.  Furthermore, the number of special solutions that go into the span are the number of free variables after elimination.  Next week we'll see that this number is called the **dimension** of the nullspace. 

Let's go back really quick and say something about the decision to continue reducing the matrix $A$ until it had an identity matrix in the columns corresponding to the basic variables.  The two reduced forms $U$ and $R$ corresponding to $A$ have special names.  The matrix $U$ is called the **echelon form** of $A$ or sometimes the **upper echelon form** of $A$.  The matrix $R$ is called the **reduced row echelon form** of $A$. 

OK, so the special solutions of $A$ live in the nullspace of $A$.  Suppose that $A$ is an $m \times n$ matrix.  In what space does $N\myp{A}$ live? 

Well, the number of entries in the special solutions correspond to the number of columns of $A$.  So the nullspace is a special set of $\mathbb{R}^n$.  

Now we're ready to state the big import fact of this section 

**Fact**: For an $m \times n$ matrix $A$, the nullspace of $A$ is a subspace of $\mathbf{R}^n$.  

To see this we need to show that vectors in the nullspace of $A$ satisfy the closure properties of a subspace.  Suppose $\bx$ and $\by$ are vectors in $N\myp{A}$: 

1. $A\myp{\bx + \by} = A\bx + A\by = {\bf 0} + {\bf 0} = {\bf 0}. \quad$ So $N\myp{A}$ is closed under vector addition. 
2. $A\myp{c\bx} = cA\bx = c\myp{A\bx} = c\myp{{\bf 0}} = {\bf 0}. \quad$ So $N\myp{A}$ is closed under scalar multiplication. 

So the nullspace of $A$ is in fact a subspace of $\mathbb{R}^n$. 

Let's close with one more example of finding the special solutions of a matrix. 

**Example 13**: Find the special solutions of $A = \mymat{1 & 2 & 1 & 3 \\ 2 & 1 & -1 & 0}$

Row reducing $A$ to row reduced echelon form, we have 

<br>

$$
\mymat{
1 & 2 & 1 & 3 \\ 
2 & 1 & -1 & 0
}
\quad \sim \quad 
\mymat{
1 & 2 & 1 & 3 \\ 
0 & -3 & -3 & -3
}
\quad \sim \quad 
\mymat{
1 & 2 & 1 & 3 \\ 
0 & 1 & 1 & 1
}
$$

<br>

$$
\sim \quad 
\mymat{
1 & 0 & -1 & 1 \\ 
0 & 1 & 1 & 1
}
$$

<br>

Writing the basic variables in terms of the free variables gives 

<br>

$$
\begin{eqnarray}
x_1 = x_3 - x_4 \\
x_2 = -x_3 - x_4 \\
x_3 = x_3 \\
x_4 = x_4 
\end{eqnarray}
\quad \Rightarrow \quad 
\bx = 
x_3 \mymat{1 \\ -1 \\ 1 \\ 0} 
+ 
x_4 \mymat{-1 \\ -1 \\ 0 \\ 1}
$$

<br>

So the special solutions are 

<br>

$$
{\bf s}_1 = 
\mymat{1 \\ -1 \\ 1 \\ 0} 
\quad \textrm{and} \quad  
{\bf s}_2 = \mymat{-1 \\ -1 \\ 0 \\ 1}
$$
