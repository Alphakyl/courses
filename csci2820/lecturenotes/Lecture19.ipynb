The **Determinant** of a square matrix $A$ is a single number denoted either $\det A$ or simply $\myab{A}$.  The determinant is a largely theoretical tool, but is essential for the discussion of eigenvalues and eigenvectors in the next lecture. 

We'll investigate the properties of determininants for simple $2 \times 2$ matrices first, and then show how to compute determinants for $n \times n$ matrices. 

The determininant of a general $2 \times 2$ matrix is given by 

<br>

$$
\det \mymat{a & b \\ c & d} = ad - bc
$$

<br>

Let's look at some examples. 

**Example 1**: 

<br>

$$
\det\mymat{1 & 0 \\ 0 & 1} = 1(1) - 0(0) = 1 
$$

<br>

$$
\det\mymat{2 & -1 \\ 3 & 5} = 2(5) - -1(3) = 13 
$$

<br>

$$
\det\mymat{0 & 1 \\ 1 & 0} = 0(0) - 1(1) = -1 
$$

<br>

$$
\det\mymat{1 & 2 \\ 2 & 4} = 1(4) - 2(2) = 0 
$$

<br>

$$
\det\mymat{a & c \\ b & d} = ad - bc 
$$

<br>

From these examples we can extrapolate some facts about determinants.  Note that while we've shown them here for the $2 \times 2$ case, all of these properties extend to the $n \times n$ case as well. 

**Fact 1**: The determinant of the identity matrix is 1: $\det I = 1$. 

<br>

**Fact 2**: The determinant of a matrix changes sign if two rows are swapped. 

<br>

**Example 2**: From Facts 1 and 2 we can state a fact about the determinant of permutation matrices.  

<br>

**Claim**: If $P$ is a permutation matrix $\det P = \pm 1$.  The determinant is $+1$ if $P$ can be turned into the identity matrix by an even number of row swaps.  The determinant is $-1$ if $P$ can be turned into the identity matrix by an odd number of row swaps. 

<br>

$$
\det \mymat{1 & 0 & 0 \\ 0 & 0 & 1 \\ 0 & 1 & 0} = -1 ~~~ \textrm{ b/c swapping rows 2 and 3 gives } ~~ I
$$

<br>

**Fact 3**: Multiplying a row of $A$ by a scalar multiplies the determinant by that scalar. 

<br>

$$
\det\mymat{\alpha a & \alpha b \\ c & d} = \alpha a d - \alpha b c = \alpha\myp{ad - bc} = \alpha \det \mymat{a & b \\ c & d}
$$

<br>

**Fact 4**: Multiplying all of $A$ by a scalar multiplies the determinant by the scalar to the $n^{th}$ power: $\det \alpha A = \alpha^n \det A$

This should be believable based on Fact 3.  If you multiply an $n \times n$ matrix $A$ by a scalar $\alpha$ this is equivalent to multiplying all $n$ rows of $A$ by $\alpha$.  Each row multiplication multiplies the determinant by $\alpha$. 

<br>

**Fact 5**: Subtracting (or adding) a multiple of one row from another does not change $\det A$. 

<br>

$$
\det\mymat{a & b \\ c - \ell a & d - \ell b} = a\myp{d-\ell b} - b\myp{c - \ell a} = ad - \ell ab - bc + \ell ab = ad - bc = \det \mymat{a & b \\ c & d}
$$

<br>

**Fact 6**: If $A$ has a zero row then $\det A = 0$. 

<br>

$$
\det \mymat{ a & b \\ 0 & 0 } = a(0) - b(0) = 0 
$$

<br>

**Fact 7**: If $A$ is triangular (upper, or lower, or diagonal) then $\det A$ is the product of the diagonal entries. 

<br>

$$
\det \mymat{a & b \\ 0 & d} = ad - b(0) = ad 
$$

<br>

**Fact 8**: The determinant of a matrix product is the product of the matrix determinants: $\det AB = \det A \det B$

<br>

$$
\det\mymat{a & b \\ c & d} \det \mymat{p & q \\ r & s} = \myp{ad-bc}\myp{ps-rq}
$$

<br>

$$
\begin{array}{rcl}
\det\mymat{ap+br & aq + bs \\ cp + dr & cq+ds} &=& (ap+br)(cq+ds)-(aq+bs)(cp+dr) \\
&=& apcq + apds + brcq + brds - aqcp - aqdr - bscp - bsdr \\
&=& apds - aqdr + brcq - bscp \\
&=& ad(ps - qr) + bc(rq - ps) \\
&=& (ad-bc)(ps-rq) \quad \checkmark
\end{array}
$$

<br>

**Fact 9**: The determinant of the transpose is unchanged: $\det A^T = \det A$

<br>

$$
\det \mymat{a & c \\ b & d} = ad - bc = \det\mymat{a & b \\ c & d}
$$

<br>

OK, here are the big ones. 

<br>

**Fact 10**: If $A$ is singular then $\det A = 0$.  If $A$ is nonsingular then $\det A \neq 0$

<br>

**Proof**:  We prove this through the LU decomposition of $A$.  Suppose we perform an LU decomposition of $A$.  If it required pivoting then we have a permutation matrix $P$, an upper triangular matrix $U$, and a **unit** lower triangular matrix $L$ such that $PA = LU$. Recall that the transpose of a permutation matrix is its inverse, so we have $A = P^TLU$. 

<br>

Then, using Facts 2, 7, 8, and 9, we have 

<br>

$$
\det A = \det P^TLU = \det P^T \det L \det U = \det P ~ \det L \det U = \pm \det U  
$$

<br>

Note that the $\pm 1$ comes from the fact that $P$ is a permutation matrix, and $\det L = 1$ because $L$ has is lower triangular and has 1's on it's main diagonal.  The remaining quantity to compute is $\det U$.  Remember that $U$ is upper triangular, and it's diagonal elements are the pivots exposed by Gaussian Elimination.  So the determinant of a matrix is equal to plus-or-minus the products of it's pivots.  

<br>

If $A$ is singular then it has a zero in a pivot position, and thus $\det A = 0$.  If $A$ is nonsingular then it has all nonzeros in pivot positions, and $\det A \neq 0$. 

$\square$

**Fact 11**: If $A$ is nonsingular then the determinant of $A^{-1}$ is equal to the reciprocal of the determinant of $A$. 

<br>

**Note**:  I've asked you to show and verify this in Homework 10. 

<br>

OK, so now we're ready to see how to compute determinants of matrices larger than $2 \times 2$.  We've already shown in the proof of Fact 10 that a valid way to compute the determinant of a matrix is to perform Gaussian Elimination on $A$, multiply the pivots of $U$ together, and then flip the sign depending on how many row swaps occured during. This is the most computationally efficient way to compute a determinant, and the method that all scientific computing platforms (i.e. Matlab, Julia, Numpy) use.  The method costs $\mathcal{O}\myp{n^3}$ operations for the row reduction.

<br>

**Example 3**: Compute the determinant of $A = \mymat{1 & 1 & 1 \\ 1 & 1 & 0 \\ 1 & 0 & 0}$.  We have 

<br>

$$
\mymat{
1 & 1 & 1 \\
1 & 1 & 0 \\
1 & 0 & 0 
}
\quad \sim \quad
\mymat{
1 & 1 & 1 \\
0 & 0 & -1 \\
0 & -1 & -1 
}
\quad \sim \quad
\mymat{
1 & 1 & 1 \\
0 & -1 & -1  \\
0 & 0 & -1 
} = U 
$$

<br> 

Then $\det A$ is the product of the diagonal entries of $U$ multiplied again by $-1$ because we performed one row swap: 

<br>

$$
\det A = (-1) \det U = (-1)[(1)(-1)(-1)] = -1
$$

<br>

**Example 4**: Compute the determinant of $A = \mymat{1 & 2 & 3 \\ 2 & 1 & 2 \\ 3 & 3 & 5}$

<br>

$$
A = 
\mymat{
1 & 2 & 3 \\ 
2 & 1 & 2 \\ 
3 & 3 & 5
}
\quad \sim \quad 
\mymat{
1 & 2 & 3 \\ 
0 &-3 & -4 \\ 
0 &-3 & -4
}
\quad \sim \quad 
\mymat{
1 & 2 & 3 \\ 
0 &-3 & -4 \\ 
0 & 0 &  0
}
$$

<br>

Then $\det A = (1)(-3)(0) = 0$, which we expect because $A$ is singular. 

<br>

A more traditional method (and the one we'll favor for computing eigenvalues) is a method called **co-factor expansion**. The basic idea is to reduce the determinant of a bigger matrix to the sum of determinants of man smaller matrices.  Consider the general $3 \times 3$ matrix along with the $3 \times 3$ *sign matrix* 

<br>

$$
\mymat{
a & b & c \\
d & e & f \\
g & h & \ell 
}
\quad \quad \quad 
\mymat{
+ & - & + \\
- & + & - \\
+ & - & + \\
}
$$

<br>

The general strategy is as follows: 

1. Choose *any* row or column of $A$ to expand along 
2. Multiply each entry in the row/column by the associate sign in the sign matrix, and the determinant of the smaller matrix obtained by blocking out the row and column containing the entry. 
3. Sum the results 

As an example, let's compute the determinant of the general $3 \times 3$ matrix above by expanding down it's first column.  We have 

<br>

$$
\det 
\mymat{
{\color{red}a} & b & c \\
{\color{red}d} & e & f \\
{\color{red}g} & h & \ell 
} = 
a\mydet{e & f \\ h & \ell} - d \mydet{b & c \\ h & \ell} + g\mydet{b & c \\ e & f}
$$

<br>

Alternatively, we could expand across the second row of the matrix

<br>

$$
\det 
\mymat{
a & b & c \\
{\color{red}d} & {\color{red}d} & {\color{red}f} \\
g & h & \ell 
} = 
-d\mydet{b & c \\ h & \ell} + e \mydet{a & c \\ g & \ell} - f\mydet{a & b \\ g & h}
$$

<br>

**Example 6**: Compute the determinant of $A = \mymat{1 & 2 & 3 \\ 2 & 1 & 2 \\ 3 & 3 & 5}$.  Expanding down the first column, we have 

<br>

$$
\begin{array}{rcl}
\det A &=& 1 \mydet{1 & 2 \\ 3 & 5} - 2 \mydet{2 & 3 \\ 3 & 5} + 3 \mydet{2 & 3 \\ 1 & 2} \\
&=& 1\myp{5 - 6} - 2\myp{10-9} + 3\myp{4-3} \\
&=& -1 -2 + 3 \\
&=& 0 \quad \checkmark
\end{array}
$$

<br>

Or, expanding across the second row, we have 

<br>

$$
\begin{array}{rcl}
\det A &=& -2 \mydet{2 & 3 \\ 3 & 5} + 1 \mydet{1 & 3 \\ 3 & 5} - 2 \mydet{1 & 2 \\ 3 & 3} \\
&=& -2\myp{10 - 9} + 1\myp{5-9} -2 \myp{3-6} \\
&=& -2 -4 + 6\\
&=& 0 \quad \checkmark
\end{array}
$$

<br>

**Example 7**: Compute the determinant of $A = \mymat{1 & 1 & 1 \\ 1 & 1 & 0 \\ 1 & 0 & 0}$

<br>

**pro-tip**: Expand over the row/column with the most zeros. 

<br>

Expanding down the third column of $A$, we have 

<br>

$$
\begin{array}{rcl}
\det A &=& 1 \mydet{1 & 1 \\ 1 & 0} - 0 \mydet{1 & 1 \\ 1 & 0} + 0 \mydet{1 & 1 \\ 1 & 1} \\
&=& 1\myp{0 - 1} \\
&=& -1 \quad \checkmark
\end{array}
$$

<br>

Co-factor expansion is convenient for small matrices, and especially small matrices with many zeros.  It will essential to our discussion of finding eigenvalues in the next lecture. 

However, it is hilariously expensive and should never be done on a computer for even remotely large matrices.  Its asymptotic cost is $\mathcal{O}(n!)$.  Computing the determinant of a $25 \times 25$ matrix (very small by modern standards) would take approximately 500,000 years on a machine that does 1 trillion flops per second!
